Cifrado cesar.

```python
import string

def encryption_cesar(message, displacement):
    uppers = string.ascii_uppercase
    lowers = string.ascii_lowercase
    encrypted_message = ""
    # Guardar abecedarios originales
    original_uppers = uppers
    original_lowers = lowers
    # Hacer el corrimiento de AMBOS abecedarios.
    displacement = displacement
    uppers = uppers[displacement:] + uppers[:displacement]
    lowers = lowers[displacement:] + lowers[:displacement]
    # Recorrer mensaje e imprimir el mensaje cifrado.
    for char in message:
        if char in original_uppers:
            index = original_uppers.index(char)
            encrypted_message += uppers[index]
        elif char in original_lowers:
            index = original_lowers.index(char)
            encrypted_message += lowers[index]
        else:
            # Si no es mayusucla ni minuscula es un caracter especial entonces respetarlo.
            encrypted_message += char
    return encrypted_message

message = input("Ingrese el mensaje a cifrar: ")
displacement = int(input("Ingrese el corrimiento: "))
print(f"Mensaje cifrado: {encryption_cesar(message,displacement)}")



```
