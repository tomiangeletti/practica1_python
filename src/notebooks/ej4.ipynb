Valide una dirección de email ingresada por el usuario con los siguientes
criterios:
Contiene exactamente un @.
Tiene al menos un carácter antes del @.
Tiene al menos un punto (.) después del @.
No empieza ni termina con @ ni con ..
La parte después del último punto tiene al menos 2 caracteres (el dominio).

```python
def is_mail(mail):
    if mail.count("@") != 1:
        return False

    if mail.startswith("@") or mail.endswith("@"):
        return False
    
    if mail.endswith(".") or mail.startswith("."):
        return False
    
    # Aca separamos el usuario del dominio.
    user, domain = mail.split("@")
    if len(user) == 0:
        return False
    if "." not in domain:
        return False
    # Validar que tenga dos caracteres despues del ultimo ".".
    parts = domain.split(".")
    if len(parts[-1]) < 2:
        return False

    return True
print("prueba")
mail = input("Ingrese el mail: ")
if is_mail(mail):
    print("Mail valido.")
else:
    print("Mail no valido.")
```
