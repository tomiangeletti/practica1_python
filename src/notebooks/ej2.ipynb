Calcule la duración total de la playlist en formato Xm Ys, y encuentre la
canción más larga y la más corta.

```python
playlist = [
{"title": "Bohemian Rhapsody", "duration": "5:55"},
{"title": "Hotel California", "duration": "6:30"},
{"title": "Stairway to Heaven", "duration": "8:02"},
{"title": "Imagine", "duration": "3:07"},
{"title": "Smells Like Teen Spirit", "duration": "5:01"},
{"title": "Billie Jean", "duration": "4:54"},
{"title": "Hey Jude", "duration": "7:11"},
{"title": "Like a Rolling Stone", "duration": "6:13"},
]

total_seconds = 0
for song in playlist:
    duration = song["duration"]
    minutes, seconds = duration.split(":")
    minutes = int(minutes)
    seconds = int(seconds)
    total_seconds += minutes * 60 + seconds

# Coversion al formato pedido.
total_minutes = total_seconds // 60
resting_seconds = total_seconds % 60
print(f"Duracion total: {total_minutes}m {resting_seconds}s")

max = 0
max_title = ""
for elem in playlist:
    duration = elem["duration"]
    minutes, seconds = duration.split(":")
    minutes = int(minutes)
    seconds = int(seconds)
    total_seconds = minutes * 60 + seconds
    if total_seconds > max:
        max = total_seconds
        max_title = elem["title"]
total_minutes = max // 60
resting_seconds = max % 60
print(f"Cancion mas larga: {max_title} ({total_minutes}m {resting_seconds}s)")

min = 999999
min_title = ""
for elem in playlist:
    duration = elem["duration"]
    minutes, seconds = duration.split(":")
    minutes = int(minutes)
    seconds = int(seconds)
    total_seconds = minutes * 60 + seconds
    if total_seconds < min:
        min = total_seconds
        min_title = elem["title"]
total_minutes = min // 60
resting_seconds = min % 60
print(f"Cancion mas corta: {min_title} ({total_minutes}m {resting_seconds}s)")
```
